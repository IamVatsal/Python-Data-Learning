```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Students Score

student_name = ['Vatsal', 'Janvi', 'Kush', 'Preet', 'Vidhi']
student_score = np.random.randint(40,100,5)
# visulize

plt.bar(student_name, student_score, color='skyblue')
plt.title("Student Score")
plt.xlabel("Student Name")
plt.ylabel("Student Score")
plt.show()
```

```python

student_score_all = np.random.randint(40,100,100)
time_studied = np.linspace(0,5,100)

plt.scatter(time_studied, student_score_all, s=student_score_all, c=student_score_all, cmap='viridis',alpha=0.7, edgecolor='k')
plt.title("Score")
plt.colorbar(label='Score Value')
plt.xlabel("Time Studiesd")
plt.ylabel("Student Score")
plt.show()
```

```python

fig, axs = plt.subplots(1,2, figsize=(10,4))

axs[0].bar(student_name, student_score, color='skyblue')
axs[0].set_title("Student Score")
axs[0].set_xlabel("Student Name")
axs[0].set_ylabel("Student Score")

axs[1].scatter(time_studied, student_score_all, s=student_score_all, c=student_score_all, cmap='viridis',alpha=0.7, edgecolor='k')
axs[1].set_title("Score")
axs[1].set_xlabel("Time Studiesd")
axs[1].set_ylabel("Student Score")

plt.suptitle("Students Score")
plt.show()
```
