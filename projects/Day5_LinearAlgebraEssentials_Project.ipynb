```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([
    [1, 0],
    [0, 1],
    [-1, 0],
    [0, -1],
    [1, 1]
])

# Plot original points
plt.scatter(points[:, 0], points[:, 1], color='blue', label='Original')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
```

```python
# Rotate by Î¸ degrees
theta = np.radians(-90)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# Scale by 1.5
scale_matrix = np.array([
    [1.5, 0],
    [0, 1.5]
])
```

```python
rotated = points @ rotation_matrix.T
scaled = rotated @ scale_matrix.T

plt.scatter(rotated[:, 0], rotated[:, 1], color='green', label='Rotated')
plt.scatter(scaled[:, 0], scaled[:, 1], color='red', label='Rotated + Scaled')

plt.legend()
plt.title("2D Transformation Simulation")
plt.grid(True)
plt.axis('equal')
plt.show()
```
