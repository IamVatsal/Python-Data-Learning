```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

student_score_all = np.random.randint(40,100,100)
student_score_all_normal = np.random.randn(100) * 10 + 75

time_studied = np.linspace(0,5,100)
subjects = ['Math', 'Science', 'English', 'History', 'Art']
avg_scores = [78, 85, 74, 80, 90]

plt.hist(student_score_all_normal, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title("Score Distribution")
plt.xlabel("Student Score")
plt.ylabel("Frequency")
plt.show()

plt.bar(subjects, avg_scores, color='orange', edgecolor='black')
plt.title("Average Scores by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.ylim(0, 100)
plt.show()

plt.scatter(time_studied, student_score_all, s=student_score_all, c=student_score_all, cmap='viridis',alpha=0.7, edgecolor='k')
plt.title("Score")
plt.colorbar(label='Score Value')
plt.xlabel("Time Studiesd")
plt.ylabel("Student Score")
plt.show()
```

```python

fig, axs = plt.subplots(2, 2, figsize=(10,8))
axs = axs.flatten()
axs[0].hist(student_score_all, bins=20, alpha=0.7, color='blue', edgecolor='black')
axs[0].set_title("Score Distribution")
axs[0].set_xlabel("Student Score")
axs[0].set_ylabel("Frequency")

axs[1].bar(subjects, avg_scores, color='orange', edgecolor='black')
axs[1].set_title("Average Scores by Subject")
axs[1].set_xlabel("Subjects")
axs[1].set_ylabel("Average Score")
axs[1].set_ylim(0, 100)

axs[2].scatter(time_studied, student_score_all, s=student_score_all, c=student_score_all, cmap='viridis',alpha=0.7, edgecolor='k')
axs[2].set_title("Score")
axs[2].set_xlabel("Time Studied")
axs[2].set_ylabel("Student Score")

# Hide the 4th subplot (axs[3])
axs[3].axis('off')
plt.tight_layout()

plt.show()
```
