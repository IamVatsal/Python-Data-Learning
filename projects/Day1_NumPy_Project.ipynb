```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.random.randint(0,15,(3,3))
print("arr1:\n", arr1)

arr2 = np.random.rand(3,3)
print("arr2:\n", arr2)
```

```python
# Sum
sum1 = np.sum(arr1)
print("\nSum of arr1: ", sum1)
print("Sum of arr1: ", arr1.sum())

sum2 = np.sum(arr2)
print("Sum of arr2: ", sum2)
print("Sum of arr2: ", arr2.sum())

sum_of_both_arr = arr1 + arr2
print("\nSum of arr1 + arr2:\n", sum_of_both_arr)
```

```python
# Mean
mean1 = np.mean(arr1)
print("\nMean of arr1: ", mean1)
print("Mean of arr1: ", arr1.mean())

mean2 = np.mean(arr2)
print("Mean of arr2: ", mean2)
print("Mean of arr2: ", arr2.mean())
```

```python
# Transpose
arr1_transpose = np.transpose(arr1)
print("\nTranspose of arr1:\n", arr1_transpose)
print("\nTranspose of arr1:\n", arr1.transpose())
print("\nTranspose of arr1:\n", arr1.T) # just returns with changed axes


arr2_transpose = np.transpose(arr2)
print("\nTranspose of arr2:\n", arr2_transpose)
print("\nTranspose of arr2:\n", arr2.transpose())
print("Transpose of arr2:\n", arr2.T) # just returns with changed axes
```

```python
plt.imshow(arr1, cmap='viridis')
plt.title("Num Array")
plt.colorbar(label='Value')
plt.show()


plt.imshow(arr2, cmap='viridis')
plt.title("Float Array")
plt.colorbar(label='Value')
plt.show()
```
