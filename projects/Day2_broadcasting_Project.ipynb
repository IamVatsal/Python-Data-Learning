```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creat np array
arr = np.random.rand(3,3)
print("Array:\n", arr)

# Visulize
plt.imshow(arr, cmap='plasma')
plt.title("Array")
plt.colorbar(label='Value')
plt.show()
```

```python
# Boolean masking to modify specific values
# Double Value < 0.5
arr_double = arr.copy()
mask = arr_double < 0.5
arr_double[mask] = arr_double[mask]*2
print("\nValue < 0.5 are Doubled:\n", arr_double)

# Visulize
plt.imshow(arr_double, cmap='plasma')
plt.title("Value < 0.5 are Doubled")
plt.colorbar(label='Value')
plt.show()

# Halve Value > 0.5
arr_halve = arr.copy()
mask = arr_halve > 0.5
arr_halve[mask] = arr_halve[mask]/2
print("\nValue > 0.5 are Halved:\n", arr_halve)

# Visulize
plt.imshow(arr_halve, cmap='plasma')
plt.title("Value > 0.5 are Halved")
plt.colorbar(label='Value')
plt.show()
```
