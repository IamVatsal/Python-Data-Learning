```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Random Data for demos

x = np.linspace(0,10,100)
y = np.sin(x)

noise = np.random.normal(0,0.1,100)
y_noise = y + noise
```

```python
# Line Plot

plt.plot(x, y, label="sin(x)")
plt.plot(x, y_noise, label="sin(x) + noise", linestyle='--')
plt.title("Line Plot Example")
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter Plot

x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(50,300,50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7, edgecolors='k')
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label='Color Intensity')
plt.show()
```

```python
# Histrogram

data = np.random.randn(1000) # Normal Distrubation

plt.hist(data, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar Chart

categories = ['Apple', 'Banana', 'Cherry', 'Mango']
values = [10,24,36,18]

plt.bar(categories, values, color=['red', 'gold', 'pink', 'orange'])
plt.title("Fruit Sales Bar Chart")
plt.xlabel("Fruit")
plt.ylabel("Sales")
plt.show()
```

```python
# Pie Chart

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("Fruit Sales Distrubation")
plt.show()
```

```python
# HEATMAP (using NumPy random data)

matrix = np.random.rand(5, 5)

plt.imshow(matrix, cmap='plasma')
plt.title("Random Matrix Heatmap")
plt.colorbar(label='Value Intensity')
plt.show()
```

```python
# SUBPLOTS (multiple plots in one figure)

x = np.linspace(0, 5, 50)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(x, y1, color='blue')
axs[0].set_title('sin(x)')

axs[1].plot(x, y2, color='red')
axs[1].set_title('cos(x)')

plt.suptitle("Subplots Example")
plt.show()
```
