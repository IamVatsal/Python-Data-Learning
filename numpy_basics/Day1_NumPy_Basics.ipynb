```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1D Array
arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr1)

# 2D Array
arr2 = np.array([[1, 2], [3, 4]])
print("2D Array:\n", arr2)

# Range of Numbers
arr3 = np.arange(0,10,2) # 0 to 10 step 2
print("Range Array:", arr3)

# Evenly Spaced Numbers
arr4 = np.linspace(0,1,5) # 5 numbers from 0 to 1
print("Linespace Array:", arr4)
```

```python
print("Shape of arr2:", arr2.shape)
print("Dimensions:", arr2.ndim)
print("Data type:", arr2.dtype)
```

```python
# Random integers
rand_ints = np.random.randint(0,10,(3,3))
print("Random integers:\n", rand_ints)

# Random floats
rand_floats = np.random.rand(3,3)
print("Random floats:\n", rand_floats)
```

```python
# Plot a 1D array
plt.plot(arr1)
plt.title("line Plot of arr1")
plt.show()

# Show 2D random floats as heatmap
plt.imshow(rand_floats, cmap='viridis')
plt.colorbar(label='Value')
plt.title("Heatmap Of Random Floats")
plt.show()
```
