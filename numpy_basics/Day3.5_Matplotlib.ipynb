```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# try 'seaborn-v0_8-darkgrid', 'fivethirtyeight', 'classic', etc.
# plt.style.use('ggplot') 
plt.style.use('seaborn-v0_8-darkgrid') 
# plt.style.use('fivethirtyeight') 
# plt.style.use('classic') 
```

```python
x = np.linspace(0, 10, 100)
y1, y2 = np.sin(x), np.cos(x)

plt.plot(x, y1, label='sin(x)', linewidth=2, linestyle='--')
plt.plot(x, y2, label='cos(x)', linewidth=2, linestyle='-.')

plt.title("Sin & Cos Waves")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.title("Annotated Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")

# Mark max point
max_x = x[np.argmax(y)]
max_y = np.max(y)
plt.scatter(max_x, max_y, color='red', zorder=5)
plt.annotate('Peak', xy=(max_x, max_y), xytext=(max_x+0.5, max_y-0.1),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1, y2, y3 = np.sin(x), np.cos(x), np.tan(x)

fig, axs = plt.subplots(3, 1, figsize=(6, 10))
axs[0].plot(x, y1, 'r')
axs[0].set_title("sin(x)")
axs[1].plot(x, y2, 'g')
axs[1].set_title("cos(x)")
axs[2].plot(x, y3, 'b')
axs[2].set_title("tan(x)")
plt.tight_layout()
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title("Distribution of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Pie chart
labels = ['A', 'B', 'C', 'D']
values = [30, 15, 45, 10]
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title("Category Distribution")
plt.show()
```
