```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b))  # 1*4 + 2*5 + 3*6 = 32
```

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print(np.dot(A, B))   # same as A @ B
print(A @ B)          # preferred in modern code
```

```python
A = np.array([[2, 3],
              [1, 4]])

det = np.linalg.det(A)
print("Determinant:", det)
```

```python
inv_A = np.linalg.inv(A)
print("Inverse:\n", inv_A)
print("\nProf:\n", inv_A @ A)
```

```python
A = np.array([[2, 3],
             [81, 4]])

B = np.array([7,7])

X = np.linalg.solve(A,B)
print("Solution (x, y):", X)
```

```python
print(np.allclose(np.dot(A, X), B))
```

```python
A = np.array([[4, -2],
              [1, 1]])

values, vectors = np.linalg.eig(A)
print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)
```
