```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4])

print(np.sqrt(arr))
print(np.exp(arr))
print(np.log(arr))
print(np.sin(arr))
print(np.power(arr,3))
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
arr = np.random.randint(1, 100, 10)
print(arr)

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Variance:", np.var(arr))
print("Min:", np.min(arr))
print("Max:", np.max(arr))
print("Sum:", np.sum(arr))                      
```

```python
matrix = np.array([[10, 20, 30],
                   [40, 50, 60],
                   [70, 80, 90]])

print(np.mean(matrix, axis=0))  # Column-wise
print(np.mean(matrix, axis=1))  # Row-wise
```

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(a + b)  # b is broadcast across rows
print(a * b)  # Element-wise multiplication with broadcasting
```

```python
arr = np.array([5, 10, 15, 20, 25])
mask = arr > 10
print(arr[mask])
# or simply
print(arr[arr > 10])
```

```python
arr = np.random.randint(1, 100, 50)
print("Array:\n",arr)

ans1 = arr[arr > 50]
print("\nfiltered Array (arr>50):\n", ans1)\

arr[arr < 30] = 0
print("\nArray with number less then 30 to zero:\n",arr)
```

```python
students = np.array([
    np.random.randint(40, 100, 4) for _ in range(5)
])

print(students)
print("Mean per student:", np.mean(students, axis=1))
print("Mean per subject:", np.mean(students, axis=0))
print("Top student index:", np.argmax(np.mean(students, axis=1)))
```
