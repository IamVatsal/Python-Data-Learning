```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Indexing Basics

arr = np.array([[10,20,30],
                [40,50,60],
                [70,80,90]])

print("Orignal Array:\n", arr)

# Access Single Element

print("\nElement at row 0, col 1:", arr[0,1])
print("Element at last row, last column:", arr[-1,-1])

# Access Entire Row or Column
print("\nFirst Row:", arr[0])
print("Second Column:", arr[:,1])
```

```python
# Slicing

print("\nTop left 2x2 subarray:\n", arr[0:2,0:2])
print("Last Two column:\n", arr[:,1:])
print("Every Second Element of fist Row:", arr[0,::2])
```

```python
# Boolean Masking

# creating a condition
mask = arr>50
print("\nBoolean Mask:\n", mask)

# appling Mask
print("\nElement Grater Than 50:", arr[mask])

# Modify Elements conditionally
arr_copy = arr.copy()
arr_copy[arr_copy<50] = 0
print("\nModified Array (Values < 50 replaced with 0):\n", arr_copy)
```

```python
# Brodcasting

# Broadcasting allows operations between arrays of different shapes
arr_b = np.array([[1,2,3]])
print("\nArray For Brodcasting:\n", arr_b)

result_add = arr + arr_b
print("\nAdding arr_b to each row:\n", result_add)

# Multiply arr by column vector
col_vactor = np.array([[1],[2],[3]])
print("\nArray For Brodcasting:\n", col_vactor)

result_mul = arr * col_vactor
print("\nMultiplying arr by column vector:\n", result_mul)
```

```python
# Visualization Example

plt.figure(figsize=(5,4))
plt.imshow(result_mul, cmap='plasma')
plt.colorbar(label='Value')
plt.title("Brodcasting Multiplication Result")
plt.show()
```
